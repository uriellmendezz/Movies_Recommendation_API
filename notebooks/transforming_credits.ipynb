```python
import pandas as pd
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

```python
credits_csv = pd.read_csv('../datasets/credits.csv')
credits_df = credits_csv.copy()
```

```python
for value in credits_df.cast[:10]:
    print(value.strip('[]'))
```

Creo nuevas columnas para alojar los datos de la columna 'cast'

```python
new_columns = ['cast_id', 'character', 'credit_id', 'gender', 'id', 'name', 'order', 'profile_path']
for column in new_columns:
    credits_df[column] = np.nan
```

```python
for index, value in enumerate(credits_df.cast[:10]):
    value = value.strip("[]").replace('cast_id','').replace('character','').replace('credit_id','').replace('gender','').replace('id','').replace('name','').replace('order','').replace('profile_path','')
    value = value.replace("'","").replace(':','')
    list_values = value.split("}")
    for v in list_values:
        v = v.replace('[','').replace(']','').replace('{','')
        v = v.split(',')
        for block in v:
            block = block.lstrip().rstrip()

            print(block)               
```
