Creo nuevas columnas para alojar los datos de la columna 'cast'

```python
import json

for index, value in enumerate(credits_df.cast[:10]):
    #new_columns = ['cast_id', 'character', 'credit_id', 'gender', 'id', 'name', 'order', 'profile_path']
    cast_id = []
    character = []
    credit_id = []
    gender = []
    id_ = []
    name = []
    order = []
    profile_path = []
    
    value = value.strip('[]')
    lista = value.split('}, {')
    for dicc in lista:
        dicc = dicc.replace('{', '').replace('}', '').replace('"', '').replace("None","'None'")
        dicc = '{' + dicc + '}'
        dicc = dicc.replace("'", '"')
        try:
            real_dict = json.loads(dicc)
            print(real_dict)
        except json.JSONDecodeError as e:
            pass
        cast_id.append(real_dict['cast_id'])
        character.append(real_dict['character'])
        credit_id.append(real_dict['credit_id'])
        gender.append(real_dict['gender'])
        id_.append(real_dict['id'])
        name.append(real_dict['name'])
        order.append(real_dict['order'])
        profile_path.append(real_dict['profile_path'])

    credits_df.at[index,'cast_id'] = str(cast_id)
    credits_df.at[index,'character'] = str(character)
    credits_df.at[index,'credit_id'] = str(credit_id)
    credits_df.at[index,'gender'] = str(gender)
    credits_df.at[index,'id'] = str(id_)
    credits_df.at[index,'name'] = str(name)
    credits_df.at[index,'order'] = str(order)
    credits_df.at[index,'profile_path'] = str(profile_path)
```
